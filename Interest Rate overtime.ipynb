```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("merged_mortgage_data.csv")
df["observation_date"] = pd.to_datetime(df["observation_date"])

plt.figure(figsize=(14, 7))

plt.plot(df["observation_date"], df["30yr_fixed_rate"], label="30-Year Fixed Rate")
plt.plot(df["observation_date"], df["15yr_fixed_rate"], label="15-Year Fixed Rate")
plt.plot(df["observation_date"], df["30yr_conforming_fico740"], 
         label="30-Year Conforming (FICO > 740)")

plt.title("Mortgage Interest Rates Over Time")
plt.xlabel("Year")
plt.ylabel("Rate (%)")
plt.legend()
plt.grid(True)

plt.show()
```
