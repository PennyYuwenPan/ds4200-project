```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("merged_mortgage_data.csv")

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="mbs_price", y="delinquency_rate_interpolated", alpha=0.6)
plt.title("Correlation: MBS Price vs. Mortgage Delinquency Rate")
plt.xlabel("MBS Price")
plt.ylabel("Delinquency Rate (%)")
plt.grid(True)
plt.show()

corr = df["mbs_price"].corr(df["delinquency_rate_interpolated"])
print("Correlation coefficient:", corr)
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df["observation_date"], df["30yr_fixed_rate"], label="30-Year Rate")

# Add crisis bands
plt.axvspan("2007-01-01", "2009-12-31", color="red", alpha=0.2, label="2008 Crisis")
plt.axvspan("2020-03-01", "2020-12-31", color="blue", alpha=0.15, label="COVID Shock")

plt.title("Mortgage Rate Timeline with Crisis Periods")
plt.xlabel("Year")
plt.ylabel("Interest Rate (%)")
plt.legend()
plt.grid(True)
plt.show()
```
